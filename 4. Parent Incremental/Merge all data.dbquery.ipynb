```sql
CREATE OR REPLACE VIEW goods_ct.gold.vw_fact_orders_enriched AS (
 SELECT 
 fo.date,
 fo.product_code,
 fo.customer_code,

 dd.date_key,
 dd.year,
 dd.month_name,
 dd.month_short_name,
 dd.quarter,
 dd.year_quarter,

 dc.customer,
 dc.market,
 dc.platform,
 dc.channel,

 dp.division,
 dp.category,
 dp.product,
 dp.variant,

 fo.sold_quantity,
 gp.price_inr,

 (fo.sold_quantity * gp.price_inr) AS total_amount_inr

 FROM goods_ct.gold.fact_orders fo

 LEFT JOIN goods_ct.gold.dim_date dd
    ON fo.date = dd.month_start_date

LEFT JOIN goods_ct.gold.dim_customers dc
    ON fo.customer_code = dc.customer_code

Left join goods_ct.gold.dim_products dp
    ON fo.product_code = dp.product_code

left join goods_ct.gold.dim_gross_price gp
    ON fo.product_code = gp.product_code
    AND YEAR(fo.date) = gp.year

);

SELECT * from goods_ct.gold.vw_fact_orders_enriched;
```
